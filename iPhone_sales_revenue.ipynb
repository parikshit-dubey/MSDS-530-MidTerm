```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Read the data from the CSV file
data = pd.read_csv('iPhone_Sales_revenue.csv')

# Print the data
print(data)
```

```python
# Extract the year, sales, and revenue columns
year = data['Year']
sales = data['Sales (mm)']
revenue = data['Revenue ($bn)']

```

```python
# Line plot of iPhone sales over the years
plt.figure(figsize=(10, 6))
plt.plot(year, sales, marker='o')
plt.xlabel('Year')
plt.ylabel('Sales (million units)')
plt.title('iPhone Sales Over the Years')
plt.xticks(year, rotation=45)
plt.grid(True)
plt.show()
```

```python
# Bar plot of iPhone revenue over the years
plt.figure(figsize=(10, 6))
plt.bar(year, revenue)
plt.xlabel('Year')
plt.ylabel('Revenue ($bn)')
plt.title('iPhone Revenue Over the Years')
plt.xticks(year, rotation=45)
plt.grid(True)
plt.show()
```

```python
# Statistical analysis
mean_sales = sales.mean()
mean_revenue = revenue.mean()
max_sales = sales.max()
max_revenue = revenue.max()

print(f"Mean sales: {mean_sales:.2f} million units")
print(f"Mean revenue: ${mean_revenue:.2f} billion")
print(f"Maximum sales: {max_sales:.2f} million units")
print(f"Maximum revenue: ${max_revenue:.2f} billion")
```

```python
# Reshape the data for modeling
X = np.array(year).reshape(-1, 1)
y_sales = np.array(sales)
y_revenue = np.array(revenue)
```

```python
# Create and fit the linear regression model for sales
sales_model = LinearRegression()
sales_model.fit(X, y_sales)
```

```python
# Predict sales using the model
sales_predictions = sales_model.predict(X)

# Create and fit the linear regression model for revenue
revenue_model = LinearRegression()
revenue_model.fit(X, y_revenue)
```

```python
# Predict revenue using the model
revenue_predictions = revenue_model.predict(X)

# Model evaluation
sales_r2 = r2_score(y_sales, sales_predictions)
revenue_r2 = r2_score(y_revenue, revenue_predictions)
```

```python
# Print the evaluations
print("===== Sales Model:")
print(f"- R-squared (R2): {sales_r2:.2f}")
print("\n")
print("===== Revenue Model:")
print(f"- R-squared (R2): {revenue_r2:.2f}")
```
